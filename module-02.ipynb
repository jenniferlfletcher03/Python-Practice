# Lesson 3

Lesson 3 is comprised of learning how to set up a dictionary in python and then pulling information from the dictionary in a print command as well as adding more keys and values to that dictionary after it is already established.

Setting up a dictionary follows a certain syntax pattern...it is similar to a list but does not function the same way that lists do. You start with a variable name:

```python
ice_cream_flavors = {}
```

and then add a structured list in curly braces that incorporates keys and values in this manner:

```python
"Mint Chocolate Chip": "Refreshing mint ice cream studded with decadent chocolate chips.",
"Cookie Dough": "Vanilla ice cream loaded with chunks of chocolate chip cookie dough.",
"Salted Caramel": "Sweet and salty with a smooth caramel swirl and a hint of sea salt."
```

The dictionary ends up being formatted like this:

```python

ice_cream_flavors = {     
    "Mint Chocolate Chip": "Refreshing mint ice cream studded with decadent chocolate chips.",
    "Cookie Dough": "Vanilla ice cream loaded with chunks of chocolate chip cookie dough.",
    "Salted Caramel": "Sweet and salty with a smooth caramel swirl and a hint of sea salt."
    }
```

Important parts to note: there needs to be a colon between the key and the value. Also, there needs to be a comma separating each key-value pair.

Because dictionaries do not function like lists, they are accessed by key and not by numeric position. A command like print(ice_cream_flavors[0]) will result in an error. In order to pull an element from the dictionary you have to know what specific element you are looking for and pull it like this:

```python
cookie_dough_description = ice_cream_flavors["Cookie Dough"]
print(cookie_dough_description)
```

or even more simply like this:

```python
print(ice_cream_flavors["Cookie Dough"])
```

You can also add things to a dictionary simply by a command like this:

```python
ice_cream_flavors["Rocky Road"] = "Chocolate ice cream mixed with other ingredients."
```

If for some reason the programmer put something wrong in the dictionary, they could use the above command to change the key or value of an exisiting element in the same way.

It is important to note that dictionaries can be composed of a mix of different data; strings, numbers, lists, etc.

# Spicy Tofu & Bell Pepper Stir-Fry

A savory, garlicky stir-fry featuring crispy tofu, sweet bell peppers, and juicy tomatoes in a tangy gluten-free sauce. This intermediate-level dish brings a moderate kick of heat balanced by the natural sweetness of the vegetables.

**Servings:** 4

## Ingredients

- 14 oz extra-firm tofu, drained and pressed
- 2 bell peppers (mixed colors), sliced
- 3 Roma tomatoes, diced
- 5 garlic cloves, minced
- 3 tbsp cornstarch
- 3 tbsp neutral oil (avocado or vegetable)
- 3 tbsp gluten-free tamari or coconut aminos
- 1 tbsp rice vinegar
- 1 tbsp maple syrup
- 1 tsp red pepper flakes
- 1 tbsp fresh ginger, grated
- 1 tsp sesame oil
- ½ tsp salt
- 2 green onions, sliced (for garnish)

## Instructions

1. **Press and cube the tofu** — Wrap the tofu in a clean towel and press for 10 minutes to remove excess water. Cut into 1-inch cubes.
2. **Coat the tofu** — Toss the tofu cubes with cornstarch and salt until evenly coated. The cornstarch creates a crispy exterior when fried.
3. **Make the sauce** — In a small bowl, whisk together tamari, rice vinegar, maple syrup, red pepper flakes, ginger, and sesame oil. Set aside.
4. **Crisp the tofu** — Heat 2 tbsp of oil in a large skillet or wok over medium-high heat. Add tofu cubes in a single layer and cook undisturbed for 3-4 minutes per side until golden and crispy. Remove and set aside.
5. **Sauté the aromatics and peppers** — Add the remaining 1 tbsp of oil to the pan. Add garlic and stir for 30 seconds until fragrant. Add bell peppers and stir-fry for 3-4 minutes until just tender-crisp.
6. **Add tomatoes** — Add the diced tomatoes and cook for 2 minutes, letting them soften slightly but keep their shape.
7. **Combine and finish** — Return the crispy tofu to the pan. Pour the sauce over everything and toss gently to coat. Cook for 1-2 more minutes until the sauce thickens and clings to the ingredients.
8. **Garnish and serve** — Remove from heat, garnish with green onions, and serve immediately over rice or rice noodles.

## Notes

Spice level: ~5/10 with 1 tsp red pepper flakes. To dial down, use ½ tsp; to dial up, add fresh chili. Press the tofu well — it's the difference between crispy and soggy. Have everything chopped and ready before you start cooking; stir-fries move fast!
