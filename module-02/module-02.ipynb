# Lesson 3

Lesson 3 is comprised of learning how to set up a dictionary in python and then pulling information from the dictionary in a print command as well as adding more keys and values to that dictionary after it is already established.

Setting up a dictionary follows a certain syntax pattern...it is similar to a list but does not function the same way that lists do. You start with a variable name:

```python
ice_cream_flavors = {}
```

and then add a structured list in curly braces that incorporates keys and values in this manner:

```python
"Mint Chocolate Chip": "Refreshing mint ice cream studded with decadent chocolate chips.",
"Cookie Dough": "Vanilla ice cream loaded with chunks of chocolate chip cookie dough.",
"Salted Caramel": "Sweet and salty with a smooth caramel swirl and a hint of sea salt."
```

The dictionary ends up being formatted like this:

```python

ice_cream_flavors = {     
    "Mint Chocolate Chip": "Refreshing mint ice cream studded with decadent chocolate chips.",
    "Cookie Dough": "Vanilla ice cream loaded with chunks of chocolate chip cookie dough.",
    "Salted Caramel": "Sweet and salty with a smooth caramel swirl and a hint of sea salt."
    }
```

Important parts to note: there needs to be a colon between the key and the value. Also, there needs to be a comma separating each key-value pair.

Because dictionaries do not function like lists, they are accessed by key and not by numeric position. A command like print(ice_cream_flavors[0]) will result in an error. In order to pull an element from the dictionary you have to know what specific element you are looking for and pull it like this:

```python
cookie_dough_description = ice_cream_flavors["Cookie Dough"]
print(cookie_dough_description)
```

or even more simply like this:

```python
print(ice_cream_flavors["Cookie Dough"])
```

You can also add things to a dictionary simply by a command like this:

```python
ice_cream_flavors["Rocky Road"] = "Chocolate ice cream mixed with other ingredients."
```

If for some reason the programmer put something wrong in the dictionary, they could use the above command to change the key or value of an exisiting element in the same way.

It is important to note that dictionaries can be composed of a mix of different data; strings, numbers, lists, etc.
