**Lesson 7**

```python
print(f"Isabel's dog age is {28/7:.0f}")
```

The :.0f function chops the decimals off of a float when it prints. You can tell python how many decimals to keep by changing the number in the function; :.1f keeps one decimal, :.2f keeps two decimals, etc.

```python
print(f"I am {2026-1982} years old.")
```

```python
print(f"There are {365/7:.0f} weeks in a year")
```

```python
print(f"The area of a square with side 5 cm is {5*5} cm squared.")
```

```python
print(f"The house was a good size: 1200 square feet, or {1200 * 0.092903:.1f} meters squared!")
```

This lesson was about using multistrings with computations as well as how f strings are displayed and how to get the decimal points you want with the :.0f command.
